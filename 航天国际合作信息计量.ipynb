### 从机构信息 抽取 国家地区组合

```python
#专用于scopus
import pycountry

def extract_countries(string):
    countries = []
    for country in pycountry.countries:
        if country.name in string:
            countries.append(country.name)
        elif hasattr(country, 'common_name') and country.common_name in string:
            countries.append(country.common_name)  #wos与scopus里都是south korea
        elif hasattr(country, 'official_name') and country.official_name in string:
            countries.append(country.official_name) #很长 一般没啥问题
    return countries

df['Countries'] = df['Affiliations'].apply(extract_countries)
df
#'Countries'这一列会出现类似['China', 'United Kingdom', 'India', 'Japan', 'Taiwan']
```

### 将国家地区组合 分成一国一行

```python
#'Countries'的['China', 'United Kingdom', 'India', 'Japan', 'Taiwan']会变成四行
def transform_dataframe(df):
    new_rows = []
    for index, row in df.iterrows():
        countries = row['Countries']
        #data = row['DATA']
        if 'China' in countries:
            for country in countries:
                if country != 'China':
                    new_row = row.copy()
                    new_row.at['partner'] = country
                    new_rows.append(new_row)
        else:
            new_rows.append(row)
    new_df = pd.DataFrame(new_rows)
    new_df.reset_index(drop=True, inplace=True)  # 重置行索引
    return new_df


df = transform_dataframe(df)
```

### 1.4分析在一国关系中的中国主导率（切片法提取第一作者的国别）

```python
df['dominance'] = 0
df['first_author'] = ''
# 迭代 DataFrame 的每一行
for index, row in df.iterrows():
    affiliations = row['Authors with affiliations']
    
    # 找到第一个分号的索引
    semicolon_index = affiliations.find(';')

    # 找到分号之前第一个逗号的索引
    comma_index = affiliations[:semicolon_index].rfind(',')

    # 提取逗号与分号之间的字符，并去除首尾空格
    extracted_string = affiliations[comma_index + 1:semicolon_index].strip()
    
    df.at[index, 'first_author'] = extracted_string
    # 判断提取的字符是否为 'China'，并设置相应的 'dominance' 值
    if extracted_string == 'China':
        df.at[index, 'dominance'] = 1

# 显示更新后的 DataFrame
df
```

```python
domi_usa = df.loc[df['partner'] == 'United States', 'dominance']
value_counts = domi_usa.value_counts()
percentages = value_counts / len(domi_usa) * 100
value_counts = value_counts.to_frame().rename(columns={'dominance': 'Frequency'})
value_counts['Percentage'] = percentages
```
